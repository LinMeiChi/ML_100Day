# 檢視與處理 Outliers
### 為何會有 outliers, 常見的 outlier 原因
* 未知值，隨意填補 (約定俗成的代入)，如年齡常見 0,999
* 可能的錯誤紀錄/手誤/系統性錯誤，如某本書在某筆訂單的銷售量 = 1000 本

```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

# 設定 data_path
dir_data = './data/'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
app_train.head()
```

## 請參考 HomeCredit_columns_description.csv 的欄位說明，觀察並列出三個你覺得可能有 outlier 的欄位並解釋可能的原因

```python
# 先篩選數值型的欄位
"""
YOUR CODE HERE, fill correct data types (for example str, float, int, ...)
"""
dtype_select = [np.dtype('int64'), np.dtype('float64')]

# isin()判斷()內的元素是否在列表中，以 bool表示。
# 將所有值的型態為 int 和 float 的欄位，包在一個list內。
numeric_columns = list(app_train.columns[list(app_train.dtypes.isin(dtype_select))])
# print("----- 資料中，所有欄位的型態 -----\n",app_train.dtypes,"\n")
# print("----- 判斷資料中，所有欄位的型態是否包含在dtype_select內 -----\n",app_train.dtypes.isin(dtype_select),"\n")
# print("----- 值的型態為int 和 float 的欄位 -----\n",numeric_columns)

# 再把只有 2 值 (通常是 0,1) 的欄位去掉
''' 
apply:把一個匿名函式 lambda 套用到 app_train[numeric_columns] 之中，並以此建立一個新的 存活 欄位。
參考:　https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.apply.html
lambda 參數1, 參數2, …: 運算式A if 關係運算式 else 運算式B。
unique: 去除重複值。
'''
numeric_columns = list(app_train[numeric_columns].columns[list(app_train[numeric_columns].apply(lambda x:len(x.unique())!=2))])
print("Numbers of remain columns %s" % len(numeric_columns))

# # 檢視這些欄位的數值範圍
for col in numeric_columns:
    # 若為下面三個檢查欄位，將印出欄位名稱變色，方便觀察。
    if  col == 'AMT_INCOME_TOTAL' or\
        col == 'REGION_POPULATION_RELATIVE'  or\
        col == 'OBS_60_CNT_SOCIAL_CIRCLE':
        print("\033[1;31;43m----- 欄位: %s ------ \n\033[0m" % col )
    else:
        print("----- 欄位: %s ------ \n" % col)
    print((app_train[col]).describe())
    plt.hist(app_train[col])
    plt.show()
```

```python
# 從上面的圖檢查的結果，至少這三個欄位好像有點可疑

# AMT_INCOME_TOTAL
# REGION_POPULATION_RELATIVE
# OBS_60_CNT_SOCIAL_CIRCLE
```

### Hints: Emprical Cumulative Density Plot, [ECDF](https://zh.wikipedia.org/wiki/%E7%BB%8F%E9%AA%8C%E5%88%86%E5%B8%83%E5%87%BD%E6%95%B0), [ECDF with Python](https://stackoverflow.com/questions/14006520/ecdf-in-python-without-step-function)

```python
# 最大值離平均與中位數很遠
print(app_train['AMT_INCOME_TOTAL'].describe())

# 繪製 Empirical Cumulative Density Plot (ECDF)
# 參考:https://medium.com/ai%E5%8F%8D%E6%96%97%E5%9F%8E/exploratory-data-analysis-%E6%8E%A2%E7%B4%A2%E8%B3%87%E6%96%99-ecdf-7fa350c32897
"""
YOUR CODE HERE
"""
cdf = app_train['AMT_INCOME_TOTAL']
cdf2 = np.sort(cdf)

plt.plot(list(cdf2), cdf.index/cdf.index.max())

plt.xlabel('Value')
plt.ylabel('ECDF')
plt.xlim([cdf.index.min(), cdf.index.max() * 1.05]) # 限制顯示圖片的範圍
plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍

plt.show()

# 改變 y 軸的 Scale, 讓我們可以正常檢視 ECDF
plt.plot(np.log(list(cdf2)), cdf.index/cdf.index.max())
plt.xlabel('Value (log-scale)')
plt.ylabel('ECDF')

plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍

plt.show()
```
