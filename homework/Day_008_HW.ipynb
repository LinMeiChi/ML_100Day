```python
##摘要統計數據
statistics = app_train["AMT_INCOME_TOTAL"].describe()
statistics
```

```python
##任意欄位採用 AMT_INCOME_TOTAL

#計算欄位的平均數
mean = app_train["AMT_INCOME_TOTAL"].mean()
print("平均數 = %f \n" % mean)

#計算欄位的標準差
std = app_train["AMT_INCOME_TOTAL"].std()
print("標準差 = %f \n" % std)
```

```python
#畫出任意欄位(AMT_INCOME_TOTAL)的直方圖
#補充: https://www.jianshu.com/p/30e4ff722cdd
# 參數介紹: https://medium.com/python-%E8%B3%87%E6%96%99%E8%A6%96%E8%A6%BA%E5%8C%96/python-%E5%95%86%E6%A5%AD%E6%95%B8%E6%93%9A%E5%88%86%E6%9E%90%E4%B9%8B%E5%8F%AF%E8%A6%96%E5%8C%96%E7%B9%AA%E5%9C%96-%E7%AC%AC7%E8%AC%9B-%E7%9B%B4%E6%96%B9%E5%9C%96-histogram-chart-6b8cd9dd26e3
app_train['AMT_INCOME_TOTAL'].hist(color='r',alpha=0.5)
plt.xlabel('AMT_INCOME_TOTAL')
```

```python
'''quantile(分位數):用分割點（cut point）將一個隨機變量的概率分布範圍分為幾個具有相同概率的連續區間。分割點的數量比劃分出的區間少1，例如3個分割點能分出4個區間。
分位數(q) : float或array-like,默認0.5(50% quantile)，
例如:quantile(0.25) -> (25,75)'''
#由於 max 和 75％ 數據差異太大，㑹看不出分佈， 故過濾最大的1％。
#參考: https://www.cupoy.com/qa/club/ai_tw/0000016D6BA22D97000000016375706F795F72656C656173654B5741535354434C5542/0000016A67EC50B00000028C6375706F795F72656C656173655155455354
#取 AMT_INCOME_TOTAL 小於四分位數(0.99)->(99,100)中，AMT_INCOME_TOTAL欄位。
app_train.loc[app_train['AMT_INCOME_TOTAL']<app_train['AMT_INCOME_TOTAL'].quantile(0.99)]['AMT_INCOME_TOTAL'].hist()
plt.xlabel('AMT_INCOME_TOTAL')
```
