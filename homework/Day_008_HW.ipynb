```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd

# 設定 data_path
dir_data = './data/'
```

```python
f_app_train = os.path.join(dir_data, 'application_train.csv')
app_train = pd.read_csv(f_app_train)
app_train
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
##摘要統計數據
statistics = app_train["AMT_INCOME_TOTAL"].describe()
statistics
```

```python
##任意欄位採用 AMT_INCOME_TOTAL

#計算欄位的平均數
mean = app_train["AMT_INCOME_TOTAL"].mean()
print("平均數 = %f \n" % mean)

#計算欄位的標準差
std = app_train["AMT_INCOME_TOTAL"].std()
print("標準差 = %f \n" % std)
```

```python
#畫出任意欄位(AMT_INCOME_TOTAL)的直方圖
#補充: https://www.jianshu.com/p/30e4ff722cdd
fig, ax = plt.subplots()

# 參數介紹: https://medium.com/python-%E8%B3%87%E6%96%99%E8%A6%96%E8%A6%BA%E5%8C%96/python-%E5%95%86%E6%A5%AD%E6%95%B8%E6%93%9A%E5%88%86%E6%9E%90%E4%B9%8B%E5%8F%AF%E8%A6%96%E5%8C%96%E7%B9%AA%E5%9C%96-%E7%AC%AC7%E8%AC%9B-%E7%9B%B4%E6%96%B9%E5%9C%96-histogram-chart-6b8cd9dd26e3
app_train['AMT_CREDIT'].diff().hist(color='r',alpha=0.5,bins=50)
plt.show()
```
