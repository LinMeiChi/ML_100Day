```python
#!pip install pandas
import os
import numpy as np
import pandas as pd
```

```python
# 設定 data_path
dir_data = './data/'
```

```python
#將目錄與文件檔(csv)結合成同一個路徑。
f_app = os.path.join(dir_data, 'application_train.csv') 
print('Path of read in data: %s' % (f_app))

#讀取csv檔案
app_train = pd.read_csv(f_app) 
```

### 如果沒有想法，可以先嘗試找出剛剛例子中提到的問題的答案
#### 資料的 row 數以及 column 數

```python
# 計算列和欄位數，回傳格式為(列數，欄位數)
app_train.shape
```

#### 列出所有欄位

```python
#列出欄位名稱
app_train.columns
```

#### 截取部分資料

```python
#loc:以 column 名來做選擇
app_train.loc[0:3,'SK_ID_CURR':'CODE_GENDER']
```

```python
#iloc:以數字為 index 來做選擇
app_train.iloc[25:33,:6]
```

#### 還有各種數之不盡的資料操作，重點還是取決於實務中遇到的狀況和你想問的問題，在馬拉松中我們也會陸續提到更多例子

```python
#選取單一 column
single_column = app_train.CODE_GENDER
print('CODE_GENDER欄位:\n', single_column)
#選取 row index
row_index = app_train[0:20]
print('\n')
print('0-19列:\n', row_index)
```

```python
# 輸出前幾筆資料
h = app_train.head()
print('\n')
print('前5筆資料:\n', h)
# 輸出後幾筆資料
t = app_train.tail(2)
print('\n')
print('前2筆資料:\n', t)
```
